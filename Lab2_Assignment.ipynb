## Question 3

```python
df = pd.read_csv('../datasets/WA_Fn-UseC_-Telco-Customer-Churn.csv',index_col='customerID')
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.info()
```

```python
df.head()
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
df.info()
```

```python
df = pd.concat([df, pd.get_dummies(df[['gender', 'Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','PaymentMethod']])], axis=1)
df.info()
```

```python
del df['gender']
del df['Partner']
del df['Dependents']
del df['PhoneService']
del df['MultipleLines']
del df['InternetService']
del df['OnlineSecurity']
del df['OnlineBackup']
del df['DeviceProtection']
del df['TechSupport']
del df['StreamingTV']
del df['StreamingMovies']
del df['PaperlessBilling']
del df['PaymentMethod']
df.info()
```

```python
ordinal_encoding_mapping = {"Month-to-month":1, "One year":2, "Two year":3}
df["Contract"] = df["Contract"].replace(ordinal_encoding_mapping)
df
```

```python
df['Contract'] = pd.to_numeric(df['Contract'], errors='coerce')
df.info()
```

```python
Bin_encoding_mapping = {"No":0, "Yes":1}
df["Churn"] = df["Churn"].replace(Bin_encoding_mapping)
df
```

```python
df['Churn'] = pd.to_numeric(df['Churn'], errors='coerce')
df.info()
```

```python
yhat_df=df['Churn']
yhat_df = pd.DataFrame(yhat_df)
yhat_df
```

```python
del df['Churn']
df
```

```python
normalized_df=(df-df.min())/(df.max()-df.min())
normalized_df
```

```python
Train_data = normalized_df.to_csv('./Train.csv', header='true')
print(Train_data)
Target = yhat_df.to_csv('./Target.csv', header='true')
print(Target)
```
